```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
data = np.array([50, 20, 20, 10, 30])
plt.pie(data)
plt.show()
```

```python
time = np.array([8, 6, 4, 2, 4])

myexplode = [0, 0.2, 0, 0, 0]
plt.pie(time, explode = myexplode )

plt.show()
```

```python
time = np.array([8, 6, 4, 2, 4])

myexplode = [0, 0.2, 0, 0, 0]
c = ["black", "gold", "pink", "red", "blue"]
plt.pie(time, explode = myexplode, colors = c )

plt.show()
```

```python
time = np.array([8, 6, 4, 2, 4])

e1 = [0, 0, 0, 0, 0]
c1 = ["black", "gold", "pink", "red", "blue"]
l1 = ["sleep", "job", "python", "game", "rest"]
plt.pie(time, explode = e1, colors = c1, labels = l1)
plt.legend(title = "time")

plt.show()
```
