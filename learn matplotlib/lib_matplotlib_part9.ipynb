```python
#add lib
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
#create data
data = np.array([10, 20, 15, 12, 35, 10, 5, 17])
job = ["python", "java", "c++", "c", "php", "js", "c#", "sql"]
e = [0.2, 0, 0, 0, 0, 0, 0, 0]

#create graph
plt.figure(figsize=(16, 8))
plt.pie(data, labels = job, autopct='%1.2f%%', explode = e)
plt.legend(title = "job")


plt.show()
```
