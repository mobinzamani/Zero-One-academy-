```python
#add lib
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#create data
x = np.random.randint(100, size = (100))
y = np.random.randint(100, size = (100))
colors = np.random.randint(100, size = (100))
size = 10 * np.random.randint(100, size = (100))
#create graph
plt.scatter(x, y, s = size, c = colors, alpha = 0.6, cmap = "ocean") #cmap = nipy_spectral / plasma / magma / winter
plt.colorbar()
plt.title("random graph", size = 20, color = "red")
plt.xlabel("x random", size = 12, color = "green")
plt.ylabel("y random", size = 12, color = "green")
plt.show()
```
