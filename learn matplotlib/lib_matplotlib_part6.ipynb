```python
#add lib
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#create data
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))

#create graph
plt.bar(x, y, color = "black")
plt.show()

```

```python
#create data
y_tick = range(1, 17)
day = range(1, 31)
time = np.random.randint(15, size=(30))      #--> or [2, 5, 4 , 9, 10, 5  ... ]

plt.figure(figsize=(12, 6))
#create graph
plt.bar(day, time, color = "green")
plt.title("python time", size = 20, color = "red")
plt.xlabel("days", size = 15, color = "blue")
plt.ylabel("time", size = 15, color = "blue")
plt.xticks(day)
plt.yticks(y_tick)
#plt.grid()

plt.show()
```
