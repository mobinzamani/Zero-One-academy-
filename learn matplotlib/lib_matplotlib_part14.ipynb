```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.show()
```

```python
plt.fill_between(x, y1, y2, color='lightblue', alpha=0.5)
plt.show()
```

```python
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.fill_between(x, y1, y2, color='green', alpha=0.5)
plt.title("Fill Between Example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
