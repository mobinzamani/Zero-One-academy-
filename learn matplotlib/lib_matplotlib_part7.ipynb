```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.array(["python", "java", "c", "php", "c++"])
y = np.array([50, 36, 51, 28, 30])

plt.bar(x , y, width = 0.6)
plt.show()
```

```python
plt.barh(x , y, color = "green", height = 0.6)
plt.show()
```

```python
x = np.array(["python", "java", "c", "php", "c++"])
y = np.array([50, 36, 51, 28, 30])

plt.bar(x, y, width = 0.6, color = ["red", "green", "blue", "black", "yellow"])

plt.title("programming", size = 25)
plt.xlabel("languages", size = 15)
plt.ylabel("point", size = 15)
plt.show()
```
